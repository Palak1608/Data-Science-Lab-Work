```python
# Given values
TP = 50
FP = 8
TN = 32
FN = 10

# Total predictions
total = TP + FP + TN + FN

# Metrics calculations
accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = (2 * precision * recall) / (precision + recall)

# Printing results
print("Total Predictions:", total)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

```python
# Given values
SSR = 420
SSE = 180

# Total Sum of Squares
SST = SSR + SSE

# R-squared calculation
R_squared = SSR / SST

# Printing results
print("Total Sum of Squares (SST):", SST)
print("R-squared value:", R_squared)

# Interpretation
print("Does the model explain more than 70% variance?", R_squared > 0.7)
```

```python
# Squared errors
errors = 2 + 4 + 1 + 9 + 4
n = 5

# MSE calculation
MSE = errors / n

# Printing result
print("Mean Squared Error (MSE):", MSE)
```

```python
# Given values
ei_squared = 9
MSE = 3
p = 3
hii = 0.2
n = 25

# Cook's Distance formula
Di = (ei_squared / (p * MSE)) * (hii / (1 - hii)**2)

# Threshold
threshold = 4 / n

# Printing results
print("Cook's Distance:", Di)
print("Threshold value:", threshold)
print("Is the observation influential?", Di > threshold)
```

```python
# Given regression equation y = 2.5x + 10
slope = 2.5
intercept = 10
x = 12

# Prediction
y = slope * x + intercept

# Printing result
print("Predicted value of y:", y)

# Explanation
print("Slope represents change in y for one unit change in x")
print("Intercept represents value of y when x is zero")
```
