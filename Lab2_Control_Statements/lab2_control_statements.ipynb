# Flatten Nested List using Python

## Description
This project demonstrates how to flatten a nested list into a single list using Python.

A nested list contains multiple sublists. The goal is to extract all elements from these sublists and combine them into one list.

---

## Dataset
- No external dataset used
- Input is a manually defined nested list

```python
Main_list = [[1,2,3],[4,5,6],[7,8,9]]

````python
Main_list = [[1,2,3],[4,5,6],[7,8,9]]

Consise_List = []
for sublist in Main_list:
    for el in sublist:
        Consise_List.append(el)

print(Consise_List)
````

````python
# Flatten Nested List

## Description
##This program converts a nested list into a single list.

## Input
Main_list = [[1,2,3],[4,5,6],[7,8,9]]

## Code
#python
Main_list = [[1,2,3],[4,5,6],[7,8,9]]

Consise_List = []
for sublist in Main_list:
    for el in sublist:
        Consise_List.append(el)

print(Consise_List)
````
